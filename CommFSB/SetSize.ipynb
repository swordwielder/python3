```python
import matplotlib.pyplot as plt

fig= plt.figure(figsize=(6,3))

axes= fig.add_axes([0.1,0.1,0.8,0.8])

x= [1,2,3,4,5]

y=[x**2 for x in x]

axes.plot(x,y)

plt.show()
```

```python
import matplotlib.pyplot as plt

fig= plt.figure(figsize=(3,6))

axes= fig.add_axes([0.1,0.1,0.8,0.8])

x= [1,2,3,4,5]

y=[x**2 for x in x]

axes.plot(x,y)

plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axes= plt.subplots(nrows=2, ncols=1,figsize=(6,3))


x= [1,2,3,4,5]

y=[x**2 for x in x]

axes[0].plot(x,y)
axes[1].plot(x,y)

plt.tight_layout()

plt.show()
```
